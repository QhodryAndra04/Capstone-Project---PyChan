```python
import pandas as pd 
import numpy as np 
```

```python
#import data
df = pd.read_excel("./datasets/data.xlsx")

df
```

```python
#retrieves all the symptom types in the symptom column
df['gejala'] = df['gejala'].apply(lambda x: x.split(', ') if isinstance(x, str) else [])

print(df['gejala'].head())
```

```python
#makes all symptom types as columns and removes the previous symptom column
gejalaList = set([gejala for sublist in df['gejala'] for gejala in sublist]) 

for gejala in gejalaList:
    df[gejala] = df['gejala'].apply(lambda x: 1 if gejala in x else 0)

df = df.drop('gejala', axis=1)

df
```

```python
#convert option text data to numeric
df['jk'] = df['jk'].map({label: idx for idx, label in enumerate(df['jk'].unique())})
df['merokok'] = df['merokok'].map({label: idx for idx, label in enumerate(df['merokok'].unique())})
df['polaMakan'] = df['polaMakan'].map({label: idx for idx, label in enumerate(df['polaMakan'].unique())})

df
```

```python
#delete columns that are not needed
df = df.drop(columns=['Timestamp']) 
df = df.drop(columns=['jk']) 
df = df.drop(columns=['tb']) 

df
```

```python
#Separate and save the data of "tidak tahu" the "asam lambung" column
testDF = df[df['asamLambung'] == 'Tidak Tahu']
testDF = testDF.drop(columns=['asamLambung']) 
testDF = testDF.reset_index(drop=True)

testDF.head()
```

```python
#collapsing rows containing the word “tidak tahu” in the column “asam lambung”
dataDF = df[df['asamLambung'] != 'Tidak Tahu']
dataDF = dataDF.reset_index(drop=True)
dataDF
```

```python
#performs label encoding to make hit or miss label data numeric
asamLambung_mapping = {
    'Ya': 1,
    'Mungkin': 1,
    'Tidak': 0
}

dataDF['asamLambung'] = dataDF['asamLambung'].map(asamLambung_mapping)

dataDF
```

```python
#save data
testDF.to_csv("./datasets/testAsamLambung.csv", index=False)
dataDF.to_csv("./datasets/datAsamLambung.csv", index=False)
```
